```python
import cv2
```

### Create a VideoCapture object and read from input file

```python
vidObj = cv2.VideoCapture('../Data/Raw/Videos/IMG_7970.MOV')
```

### Check if video file is opened successfully

```python
if (vidObj.isOpened()== False):
    print("Error opening video file")
```

### Read counter from config file

```python
counter = 0
with open('../miscFiles/Count.txt', 'rb') as inp:
    counter = int(inp.read())
    
print (counter)
```

### Create Frames directory if not present

```python
import os
path = "../Data/Raw/Frames"
if os.path.exists(path) == False :
    mode = 0o666
    os.mkdir(path, mode)
    print("Directory '% s' created" % path)
```

### Check frames per second for the video

```python
fps = vidObj.get(cv2.CAP_PROP_FPS)

# For OpenCV 3, you can also use the following
# double fps = video.get(CAP_PROP_FPS);

print ("Frames per second using video.get(cv2.CAP_PROP_FPS) : {0}".format(fps))
```

### Cehck if frames are extracted 

```python
success = 1
frame = 1
while success:
    success, image = vidObj.read()
    # Saves the frames with frame-count
    if frame%10 == 0:
        print("Saving frame frame_%d.jpg. " % counter)
        cv2.imwrite("../Data/Raw/Frames/frame_%d.jpg" % counter, image)
        counter += 1
    frame += 1
```

### Write the counter back to the config file

```python
with open('../miscFiles/Count.txt', 'w') as outp:
    count = str(counter)
    outp.write(count)
```

### Release Video

```python
vidObj.release()
print ("Done!")
```

# OR
### Read all video files at once and start extracting the frames per second out of the video being read.

```python
counter = 0
with open('../miscFiles/Count.txt', 'r') as inp:
    counter = int(inp.read())
    
listDir = "../Data/Raw/Videos" 
listing = os.listdir(listDir)
for vid in listing:
    vid = listDir + "/" +vid
    vidObj = cv2.VideoCapture(vid)
    if (vidObj.isOpened()== False):
        print("Error opening video file")
    success = True
    frame = 1
    while success:
        success, image = vidObj.read()
        # Saves the frames with frame-count
        if frame%30 == 0:
            print("Saving frame frame_%d.jpg. " % counter)
            cv2.imwrite("../Data/Raw/Frames/frame_%d.jpg" % counter, image)
            counter += 1
        frame += 1
        
```

```python
with open('../miscFiles/Count.txt', 'w') as outp:
    #count = str(counter)
    outp.write(str(counter))
    
vidObj.release()
print ("Done!")
```
