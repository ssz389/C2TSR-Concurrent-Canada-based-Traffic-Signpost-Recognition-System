```python
plt.figure(figsize=(15,15))
df["class_name"].value_counts().plot.bar()

plt.show()
```

```python
plt.figure(figsize=(20,10))
df["id"].value_counts().plot.pie()
plt.gca().set_aspect("equal")
```

```python
labels = []
plt.figure(figsize=(20,20))
for i, dfi in enumerate(df.groupby(["id"])):
    labels.append(dfi[0])
    plt.bar(i, dfi[1].count(), label=dfi[0])
plt.xticks(range(len(labels)), labels)
plt.legend()
plt.show()
```

```python
s = df['class_name'].value_counts()
print (s)
plt.figure(figsize=(20,10))
ax=s.plot.bar(width=.8) 
for i, v in s.reset_index().iterrows():
    ax.text(i, v.class_name + 0.2 , v.class_name, color='red')
```
